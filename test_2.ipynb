```python
import numpy as np
import matplotlib.pyplot as plt
import k3d
data = np.loadtxt('output_test_case_2A.csv', delimiter=',')
x = data[:,0]
y = data[:,1]
z = data[:,2]

vertices = np.vstack([x,y,z]).T
plt_line = k3d.line(vertices, width=0.1)

plot = k3d.plot()
plot += plt_line
plot.display()
```

```python
data_1 = np.loadtxt('output_test_case_2A_error.csv', delimiter=',')
data_2 = np.loadtxt('output_test_case_2B_error.csv', delimiter=',')
data_3 = np.loadtxt('output_test_case_2C_error.csv', delimiter=',')
data_4 = np.loadtxt('output_test_case_2D_error.csv', delimiter=',')
data_5 = np.loadtxt('output_test_case_2E_error.csv', delimiter=',') 
plt.plot(data_1[:,0], data_1[:,1], label='A', linestyle = "--")
plt.plot(data_2[:,0], data_2[:,1], label='B', linestyle = "--")
plt.plot(data_3[:,0], data_3[:,1], label='C', linestyle = "--")
plt.plot(data_4[:,0], data_4[:,1], label='D', linestyle = "--")
plt.plot(data_5[:,0], data_5[:,1], label='E', linestyle = "--")
plt.legend()
plt.xlabel("t/M")
plt.ylabel("$\\frac{|r-r_0|}{r}$")
plt.title("Error of orbits")
plt.savefig("Error of orbits.png")
```
